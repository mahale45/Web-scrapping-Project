```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import pyuser_agent
```

```python
ua = pyuser_agent.UA()
```

```python
url='https://www.amazon.in/s?k=mobile&crid=G7XE5FUHQLA2&sprefix=%2Caps%2C212&ref=nb_sb_ss_recent_2_0_recent'
```

```python
headers = {"User-Agent" : ua.random}
```

```python
page=requests.get(url,headers=headers)
```

```python
page
```

```python
soup=BeautifulSoup(page.text,'html.parser')
```

```python
soup
```

```python
Product_names=[]
Amount=[]
Price=[]
```

```python
names=soup.find_all('span',class_='a-size-medium a-color-base a-text-normal')
for i in names:
    name=i.text
    Product_names.append(name)
print(Product_names)
print(len(Product_names))


```

```python
amount=soup.find_all('span',class_='a-price-whole')
for i in amount:
    am=i.text
    Amount.append(am)
print(Amount)
print(len(Amount))



    
```
